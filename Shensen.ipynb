## Import tools

```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv("Shensen Waste.csv")

df.head()
```

```python
df.info()
```

```python
name_cols = ['FirstName', 'LastName']
```

```python
#Remove irregular characters from the 'FirstName', 'LastName' and convert all entries in the 'FirstName' column to lowercase for consistency
df[name_cols] = df[name_cols].apply(lambda x: x.astype(str).str.lower().str.replace(r'\s+', ' ', regex=True).str.strip())

# Print a confirmation message to indicate the conversion has taken place
print("Converted 'name_cols' to lowercase.")

# Display the first few rows of the DataFrame to show the updated 'FirstName' column
df.head()
```
